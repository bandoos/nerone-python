```python
from typing import List
import pandas as pd

import xmltodict
import os
import re

NERONE = "/home/bandoos/repos/nerone_mvn"
file = os.path.join(
    NERONE,
    "test/evaluation/resources/corpora/en/CoNNL2003/train/control/eng.train.control.united.xml",
)

line_re = r"\[(\d+)] (.+) \[(\d+)\]$"
_line_re = re.compile(line_re)


def read_xml_item(path):
    with open(path, "rb") as f:
        data = xmltodict.parse(f)
        item = data["rss"]["channel"]["item"]
        return item


# with open(file, "rb") as f:
#     data = xmltodict.parse(f)

# item = data["rss"]["channel"]["item"]

item = read_xml_item(file)

# print(len(item["ORGANIZATION"]))

# example = item["ORGANIZATION"][0]

# start, text, end = re.match(line_re, example).groups()


def parse_tag_content(line):
    r = _line_re.match(line)
    if r is not None:
        start, text, end = r.groups()
        return (int(start), int(end), text)


# orgs = list(map(parse_tag_content, item["ORGANIZATION"]))


def tabulate(item, tags: List[str]):
    records = []
    for tag in tags:
        conts = map(parse_tag_content, item[tag])
        conts = filter(lambda x: x, conts)
        with_tag = map(lambda xs: (tag, *xs), conts)
        records.extend(with_tag)

    return pd.DataFrame(records, columns=["tag", "start", "end", "text"])


df_ground = tabulate(item, ["ORGANIZATION", "PERSON", "LOCATION"])

#df.to_csv("./evaluation-set-test.csv", index_label="index")
```

```python
from itertools import islice
from tqdm import tqdm
```

```python
FIRST_N=len(df_ground)
PERFECT_MATCH=0
AREA_MATCHES_TEXT_ERROR=[]
TAG_ERRORS=[]
MISS=[]

MATCHED=0
for i, row in tqdm(islice(df_ground.iterrows(),FIRST_N),total=FIRST_N):
    #print(f"ground[{row.start}\n",row)
    #print(df[df.start == row.start])
    match = df[df.start == row.start]
    #print(len(match))
    if len(match) == 0:
        MISS.append(row)
        continue
        
    _m = match.iloc[0]
    end = _m["end"]

    ref_text = row["text"].lower()
    pred_text = _m["text"].lower()

    # then the area matches
    if row.end == end:
        MATCHED
        if ref_text == pred_text:
            PERFECT_MATCH += 1
            continue
        # span ok but text incorrect
        if ref_text != pred_text:
            AREA_MATCHES_TEXT_ERROR.append((row, _m))

        # span ok but
        if row["tag"] != _m["tag"]:
            TAG_ERRORS((row,_m))

        
    #print(f"match[{row.start}]\n", match)

    #print("---"*10)
```

```python
def df_to_prodigy_spans(df: pd.DataFrame):
    spans = []
    for _,row in df.iterrows():
        spans.append({"label":row.tag, "start":row.start, "end":row.end})
    return [spans]
```

```python
def my_conll_to_spans(doc):

    out = []

    doc = split_list(doc.split("\n"))

    for example in doc:
        labels = []
        #print(example)
        for token in example:
            #print(token)
            token = token.split(" ")
            print(token)
            label = token[-1]
            labels.append(label)
        out.append(labels)

    spans = list_to_spans(out)

    return spans
```
